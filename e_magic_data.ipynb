```python
import pandas as pd
```

```python
# Read the data from a file

import os
import shutil
dataframe = []

# Set the path to the folder containing MSG files
folder_path = r"C:\Users\Ratan Kumar Jha\Desktop\e magic data\csv_data"
Output_path = r"C:\Users\Ratan Kumar Jha\Desktop\e magic data\OUTPUT DATA"
# -------------------------------------------------------------------------------------------------
for filename in os.listdir(folder_path):
    file_path_name = os.path.join(folder_path, filename)
    data_load = pd.read_csv(file_path_name)
    print(filename)
    dataframe.append(data_load)

concatenated_dataframe = pd.concat(dataframe, ignore_index=True)

output_csv_path = os.path.join(Output_path, "load_server_concated_Data.csv")

# Save the concatenated DataFrame to a CSV file
concatenated_dataframe.to_csv(output_csv_path, index=False)

print("Concatenation completed and saved to:", output_csv_path)
```
