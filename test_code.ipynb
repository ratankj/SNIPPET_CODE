```python
FINAL_DATA_ALL_MERGE['Aging_Creation_Date'] = pd.to_datetime(FINAL_DATA_ALL_MERGE['Aging_Creation_Date'])
```

```python
FINAL_DATA_ALL_MERGE.drop(['Section ID','WO Status','Work Order Type',
        'S2S Meter',
       'CA_NMI', 'Len', 'Duplicate', 'MDM_Status', 'Never', 'HES',
       'NMI Status', 'mdm_AMR', 'mdm_master_data_sync',
       'mdm_Notification', 'Commissioning_Tag', 'Remarks','Mobile Number',
       'Type', 'Phase', 'Town',
       'Origin_ID', 'Section_Name', 'Division_Name', 'Status',
       'Vendor_Planner', 'Refurbished', 'Diff', '2023-06-17 00:00:00',
       'Refirbished', '2024-06-17 00:00:00','Communicating','Meter Replacement','Unnamed: 0'],axis=1, inplace=True)
```

```python
# Add lead status and lead date columns
FINAL_DATA_ALL_MERGE['lead_status'] = FINAL_DATA_ALL_MERGE.groupby('New Meter No')['Detailed_Remarks'].shift(-1)
FINAL_DATA_ALL_MERGE['lead_date'] = FINAL_DATA_ALL_MERGE.groupby('New Meter No')['Aging_Creation_Date'].shift(-1)

# Fill NaN values in lead_status and lead_date with current status and date
FINAL_DATA_ALL_MERGE['lead_status'].fillna(FINAL_DATA_ALL_MERGE['Detailed_Remarks'], inplace=True)
FINAL_DATA_ALL_MERGE['lead_date'].fillna(FINAL_DATA_ALL_MERGE['Aging_Creation_Date'], inplace=True)

# Calculate aging of status
FINAL_DATA_ALL_MERGE['aging'] = (FINAL_DATA_ALL_MERGE['lead_date'] - FINAL_DATA_ALL_MERGE['Aging_Creation_Date']).dt.days.fillna(1)
FINAL_DATA_ALL_MERGE['aging'] = FINAL_DATA_ALL_MERGE['aging'].where(FINAL_DATA_ALL_MERGE['aging'] > 0, 1)



#print(FINAL_DATA_ALL_MERGE)
```
